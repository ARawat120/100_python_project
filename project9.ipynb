```python
# What are tuples?

# A tuple in Python is an ordered, immutable collection of items. It is one of Python's built-in data types used to store collections of data, similar to lists, but with the key distinction of immutability.
# Key characteristics of Python tuples:
# Ordered: The items in a tuple have a defined order, and this order is preserved. Items are indexed starting from 0.
# Immutable: Once a tuple is created, its contents cannot be changed. You cannot add, remove, or modify elements within a tuple after its creation.
# Allows Duplicate Values: Since tuples are indexed, they can contain items with the same value.
# Heterogeneous: Tuples can store elements of different data types within the same tuple.


tuple1 = ('hello','mother',1999,45)
tuple2 = ('good','morning')
print(tuple1[0])
print(tuple1[-1])
```

```python
# Tuples Operations and Unpacking
print(len(tuple1))     # length of tuple

print(tuple1+tuple2)    # Concatenation

coordinates = (0,3,9)
x,y,z = coordinates      # Tuple unpacking
print(x)
print(y)
print(z)
```

```python
# What are Sets?

# set is an unordered collection of unique, immutable elements. It is one of Python's built-in data types, alongside lists, tuples, and dictionaries.
# Here are the key characteristics of a Python set:
# Unordered: Elements within a set do not have a defined order, and their order may change. You cannot access elements by index.
# Unique: A set cannot contain duplicate elements. If you try to add a duplicate, it will be ignored. This makes sets useful for removing duplicates from other collections.
# Mutable (for the set itself): You can add or remove elements from a set after it has been created.
# Immutable Elements: While the set itself is mutable, the elements stored within a set must be immutable (e.g., numbers, strings, tuples). You cannot store mutable objects like lists or dictionaries directly within a set.



set1 = {'Hello','Gabbar',5423,1000.00}
print(set1)
set1.add('Groom')    # addding element in set
print(set1)
set1.remove('Hello')    # removing element in set
print(set1)
```

```python
# Set Operaton(Union,Intersection,Difference)

set1 = {'Raman','Ishaan','Prashant','Ujjwal','Bharat','Aryan'}
set2 = {'Anushal','Raman','Aman','Prashant'}
print(set1.union(set2))     # union (|)
print(set1.intersection(set2))      #intersection (&)
print(set1.difference(set2))      # difference (-)
```
