```python
# what are if-else statement.
# odd-even.

num = int(input("Enter the number: ",))
if num%2==0:
    print(f"{num} is Even number..")
else:
    print(f"{num} is odd number")

```

```python
# comparison operator.
# voting eligibility

age = int(input("Enter your age : ",))
if age>=18:
    print("You eligible for voting...")
else:
    print("You not eligible for voting....")


```

```python
# Logical operator.
# grading system

marks = int(input("Enter your marks : ",))
if marks > 90:
    print("A")
elif marks > 80 or marks <= 90:
    print("B")
elif marks <= 80 or marks > 70:
    print("C")
elif marks > 50 or marks <= 70:
    print("D")
elif marks <= 50 or marks > 33:
    print("E")
else:
    print("F")
```

```python
# nested if-else statement.

# number must be even and also divisible by 3.
import cmath
value = int(input("Enter a number: ",))
if value%2==0:
    if value%3==0:
        print(f"{value} is even and also divisible by 3.....")
    else:
        print("value is not divisible by 3..")
else:
    print("number is not Even..")
```
