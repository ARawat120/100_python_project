```python
# What are Dictionary?
# A Python dictionary, also known as a dict, is a built-in data type that stores data as a collection of key-value pairs. It is a highly efficient and versatile data structure, often referred to as a hash table or associative array in other programming languages.
# Key Characteristics:
# Key-Value Pairs: Dictionaries store data in pairs, where each unique "key" is associated with a specific "value."
# Curly Braces: Dictionaries are defined using curly braces {}. Key-value pairs within the braces are separated by commas, and the key and value within each pair are separated by a colon (:).
# Keys Must Be Unique and Immutable: Each key within a dictionary must be unique and immutable (e.g., strings, numbers, tuples). Mutable types like lists or other dictionaries cannot be used as keys.
# Values Can Be Any Type: Values in a dictionary can be of any data type and can be duplicated.
# Ordered (from Python 3.7 onwards): While historically dictionaries were unordered, from Python 3.7 onwards, dictionaries maintain the insertion order of their key-value pairs.
# Changeable: Dictionaries are mutable, meaning you can add, remove, or modify key-value pairs after creation.

dict1 = {
    'Aman': 11,
    'Rahul': 49,
    'Ishaan':930
}
```

```python
# Accessing and Modifying Dictionary Values

print(dict1)

print(dict1['Ishaan'])   #accessing
print(dict1.get("Aman"))


dict1['Raj'] = 89
print(dict1)
```

```python
# Adding and Removing Enteries
dict1['Rajesh'] = 890    # adding
print(dict1)

del dict1['Rahul']
print(dict1)

dict1.pop('Aman')
print(dict1)
```

```python
# Looping Through a Dictionary
for key,value in dict1.items():
    print(key,value)
```
