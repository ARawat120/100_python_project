```python
# what is list??
#  a list is a versatile and fundamental built-in data type used to store an ordered collection of items. Lists are defined by enclosing comma-separated items within square brackets [].
# Here are key characteristics and common operations related to Python lists:
# Characteristics:
# Ordered: List items have a defined order, and this order is maintained. New items are typically added to the end.
# Changeable (Mutable): Lists can be modified after creation. You can add, remove, or change items.
# Allow Duplicates: Lists can contain multiple items with the same value.
# Indexed: Items are accessed using numerical indices, starting from 0 for the first item.
# Heterogeneous: Lists can store items of different data types (e.g., integers, strings, booleans, other lists).


breakfast_list = ["Milk","Bread","Jam","Butter"]
print(breakfast_list)
```

```python
# List Operation: Adding,Removing,Accessing items...
breakfast_list.append("Egg")
breakfast_list.insert(0,"Juice")# Adding Element
print(breakfast_list)
breakfast_list.remove("Jam")  # Removing Element
print(breakfast_list)
breakfast_list.pop(3)
print(breakfast_list)
print(breakfast_list[1])   # Accessing Elemnt
```

```python
# Looping Through Lists...
for item in breakfast_list:
    print(item)
```

```python
# List Methods...

# List Methods
# Let's look at different list methods in Python:
#
# append(): Adds an element to the end of the list.
# copy(): Returns a shallow copy of the list.
# clear(): Removes all elements from the list.
# count(): Returns the number of times a specified element appears in the list.
# extend(): Adds elements from another list to the end of the current list.
# index(): Returns the index of the first occurrence of a specified element.
# insert(): Inserts an element at a specified position.
# pop(): Removes and returns the element at the specified position (or the last element if no index is specified).
# remove(): Removes the first occurrence of a specified element.
# reverse(): Reverses the order of the elements in the list.
# sort(): Sorts the list in ascending order (by default).
```

```python
# Shopping List App...

Shopping_list = []

while True:
    choice = int(input(
      "1. for view the shopipng list "
      "2. for adding item "
      "3. for updating item "
      "4. for removing item "
      "5. for clear shopping list "
      "6. for exit "
      ))

    if choice == 1:
        print("items in the list are.. ",Shopping_list)

    elif choice == 2:
        item = input("Enter the product name: ",)
        Shopping_list.append(item)
        print("adding list: ",Shopping_list)

    elif choice == 3:
        item = input("Enter the updated item ")
        remove_index = int(input("Enter the modify index : ",))
        Shopping_list[remove_index] = item
        print("updated list: ",Shopping_list)

    elif choice == 4:
        item = input("Enter the remove item: ",)
        Shopping_list.remove(item)
        print("removing items from list: ",Shopping_list)

    elif choice == 5:
        break;
        print("Thanks for Shopping")

    else:
        print("Plsss Enter the correct Choice ...")

```
