```python
# Simple Calculator Project.

print(": : Welcome here for calculating : : ")
num = int(input("Enter the first value ",))
num1 = int(input("Enter the first value ",))

task = int(input("Enter the number for particular task...   "
            "1. Addition. "
            "2. Subtraction. "
            "3. Multiplication. "
            "4. Division. "
            "5. Remainder. "
            "6. Floor Division. "
              ))
if task == 1:
    print(num+num1)
elif task == 2:
    print(num-num1)
elif task == 3:
    print(num*num1)
elif task == 4:
    print(num/num1)
elif task == 5:
    print(num%num1)
elif task == 6:
    print(num//num1)
else:
    print("Plss enter number")
print("Thnk for using my code")
```
