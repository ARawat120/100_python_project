```python
# What is Loops?
# Loops use in programming for repetitive work(like counting from 1 to 100).

```

```python
# For Loops.
print("Display Whole number...")
for i in range(0,100):
    print(i)
```

```python
# While Loops.
print("Display Alphabet Series")
i=0
a ='a'
while i<27:
    print(chr(ord(a)+i))
    i=i+1
```

```python
#Using time.sleep() for delays.
import time
for i in range(5,0,-1):
    time.sleep(3)
    print("Happy Birthday")
```

```python
# Countdown Timer.

import time

start = int(input("Enter a number to start Countdown from : ",))

print("-------Countcown--Begin-------")
while start>0:
    time.sleep(1)
    start-=1
print("Happy Newyear....")
```
